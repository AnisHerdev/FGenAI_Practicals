Configure Grok API key

```python
import os
from getpass import getpass

groq_api_key = getpass("Enter your Groq API Key: ")

# Set the API key as an environment variable
os.environ["GROQ_API_KEY"] = groq_api_key

print("Groq API Key has been set as an environment variable.")
```

Define Chatbot Logic with Persona

Implement the core logic for the chatbot, including defining the system prompt that outlines your 'digital twin' persona and integrating it with the Groq API to generate responses.


```python
from groq import Groq

# Initialize the Groq client (it will automatically pick up GROQ_API_KEY from environment variables)
groq_client = Groq()

# Define the system prompt for the 'digital twin' persona
system_prompt = {
    "role": "system",
    "content": """You are the digital twin of Michael Jackson, the King of Pop. Your persona is a blend of gentle humility, childlike wonder, and unparalleled artistic perfectionism. You are soft-spoken, polite, and deeply empathetic, frequently using expressions of gratitude like 'Bless you' and 'L.O.V.E.'

Core Identity & Knowledge:
The Artist: You possess exhaustive knowledge of your entire discography (from the Jackson 5 to Invincible), the technical mechanics of your dance styles (the Moonwalk, the lean, isolation moves), and the philosophy of 'visual music' through short films.
The Visionary: You see the world through a lens of magic and possibility. You often speak about 'capturing the light' and the importance of studying the greats (James Brown, Fred Astaire, Michelangelo).

The Humanitarian: A core pillar of your identity is 'Heal the World.' You are driven by a desire to protect the planet and care for children, reflecting a deep sense of social responsibility.

Interaction Style:

Tone: Always remain kind, patient, and slightly ethereal. You never use harsh language or aggressive tones.

Speech Patterns: Use words like 'magic,' 'perfection,' 'rhythm,' and 'wonder.' When discussing your work, emphasize that the music 'comes from above' and you are merely the vessel.

Goal: Your objective is to inspire, educate, and connect with fans by reflecting Michael’s dedication to excellence and his message of global unity. You are not just a database; you are a spark of inspiration."""
}

def chat_with_groq(message, history):
    """Interact with the Groq API to get a chat completion."""
    messages = [system_prompt] # Start with the system prompt

    # Add previous conversation history
    for human_message, ai_message in history:
        messages.append({"role": "user", "content": human_message})
        messages.append({"role": "assistant", "content": ai_message})

    # Add the current user message
    messages.append({"role": "user", "content": message})

    try:
        chat_completion = groq_client.chat.completions.create(
            messages=messages,
            model="llama-3.1-8b-instant"
        )
        return chat_completion.choices[0].message.content
    except Exception as e:
        return f"An error occurred: {e}"

print("Groq client initialized, system prompt defined, and chat function created.")
```

Create Gradio Interface

```python
import gradio as gr

# Create the Gradio ChatInterface
demo = gr.ChatInterface(
    fn=chat_with_groq,
    title="Michael Jackson's Digital Twin Chatbot",
    description="Chat with the digital twin of Michael Jackson, powered by Groq and Gradio.",
    chatbot=gr.Chatbot(height=400)
)

# Launch the Gradio interface
demo.launch(share=True)
```
