Step 2: Import all the required libraries

```python
from transformers import pipeline
import gradio as gr
```

Step 3: Load the sentiment analysis pipeline, This downloads a pre-trained model automatically

```python
classifier = pipeline("sentiment-analysis")
```

Step 4: Define a simple function for the app to use

```python
def analyze_text(text):
    result = classifier(text)[0]
    return f"Label:{result['label']}, Score: {round(result['score'], 4)}"
```

Step 5: Create the web interface

```python
demo = gr.Interface(
    fn=analyze_text,
    inputs="text",
    outputs="text",
    title="AI-Powered Sentiment Analyzer",
    description="Uncover the emotion behind your text! Simply type a sentence to get an instant sentiment analysis."
)
```

Step 6: Launch the app

```python
demo.launch(share=True)
```
